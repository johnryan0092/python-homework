```python
# From the pathlib and csv library
from pathlib import Path
import csv

#check the current directory from where the Python program is executing from
#print(f"Current Working Directory: {Path.cwd()}")

#set the file path
csvpath = Path("../PyBank/budget_data.csv")

#open the csv file as an object
with open(csvpath, "r") as csvfile:
    
    
    #pass the csvfile in csvreader
    # with ',' as delimiter
    csvreader = csv.reader(csvfile, delimiter = ",")
    
    #read the header row
    csv_header = next(csvreader)
    #print the header
    #print(csv_header)

    #initialize variables
    row = next(csvreader)
    max_pnl_month = row[0]
    min_pnl_month = row[0]
    pnl = int(row[1])
    max_pnl = pnl
    min_pnl = pnl
    previous_pnl = pnl
    total_pnl = int(row[1])
    total_pnl_change = 0
    total_months = 1



    
    #read each row after header
    for row in csvreader:
        
        #print row
        #print(row)
        
        
        pnl = int(row[1])
                  
        
        
        #count total months         
        total_months = total_months + 1
                  
        #summarize pnl
        total_pnl = total_pnl + pnl
        #pnl change
        pnl_change = pnl - previous_pnl
        #summarize the total pnl changes for dataset
        total_pnl_change = total_pnl_change + pnl_change
                  
        #min/max determination
        if pnl_change > max_pnl:
                  max_pnl_month = row[0]
                  max_pnl = pnl_change
        
        if pnl_change < min_pnl:
                  min_pnl_month = row[0]
                  min_pnl = pnl_change
                  
        #define previous pnl
        previous_pnl = pnl
                
                  
    #Calculations
    average_pnl = total_pnl / total_months
                  
    average_pnl_change = total_pnl_change / (total_months -1)
    
    #Formatting
    total_pnl = round(total_pnl, 2)
    average_pnl_change = round(average_pnl_change, 2)
    max_pnl = round(max_pnl, 2)
    min_pnl = round(min_pnl, 2)
    
                  
                  
    #Summary
    print(f"Financial Analysis:")
    print(f"---------------------------------------------")
    print(f"Total Months:  {total_months}")
    print(f"Total PnL:   ${total_pnl}")
    print(f"Average PnL Change: ${average_pnl_change}")
    print(f"The greatest increase in PnL took place {max_pnl_month} amounting to ${max_pnl}.")
    print(f"The greatest decrease in PnL took place {min_pnl_month} amounting to ${min_pnl}.")
    
        
    #set output file
    output_path = 'output.txt'
    #open the outputfile and print results
    with open(output_path, 'w') as file:
        file.write(f"Financial Analysis:\n")
        file.write(f"---------------------------------------------\n")
        file.write(f"Total Months:  {total_months}\n")
        file.write(f"Total PnL:   ${total_pnl}\n")
        file.write(f"Average PnL Change: ${average_pnl_change}\n")
        file.write(f"The greatest increase in PnL took place {max_pnl_month} amounting to ${max_pnl}.\n")
        file.write(f"The greatest decrease in PnL took place {min_pnl_month} amounting to ${min_pnl}.")
        

        
        
        

```
